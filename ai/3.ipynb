### Библиотеки

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from sklearn.neighbors import KNeighborsClassifier
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
```

### Задание 1.3.1

```python
x = np.array([0, 0, 0])
y = np.array([1, 1, 1])
z = np.array([1, 1, 0])
w = np.array([0, 1, 1])

# Расстояние Евклида
print(np.linalg.norm(x - y))
print(np.linalg.norm(x - z))
print(np.linalg.norm(x - w))
print(np.linalg.norm(w - y))
print(np.linalg.norm(w - z))
print(np.linalg.norm(y - z))

# Квадрат Евлидова расстояния
print(np.linalg.norm(x - y)**2)
print(np.linalg.norm(x - z)**2)
print(np.linalg.norm(x - w)**2)
print(np.linalg.norm(w - y)**2)
print(np.linalg.norm(w - z)**2)
print(np.linalg.norm(z - y)**2)

# Расстояние Чебышева
print(np.linalg.norm(abs(x - y), ord=np.inf))
print(np.linalg.norm(abs(x - z), ord=np.inf))
print(np.linalg.norm(abs(x - w), ord=np.inf))
print(np.linalg.norm(abs(w - y), ord=np.inf))
print(np.linalg.norm(abs(w - z), ord=np.inf))
print(np.linalg.norm(abs(z - y), ord=np.inf))

# Расстояние Хемминга
print(np.linalg.norm(abs(x - y), ord=1))
print(np.linalg.norm(abs(x - w), ord=1))
print(np.linalg.norm(abs(x - z), ord=1))
print(np.linalg.norm(abs(y - z), ord=1))
print(np.linalg.norm(abs(y - w), ord=1))
print(np.linalg.norm(abs(w - z), ord=1))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(0, 0, 0) # x
ax.scatter(1, 1, 1) # y
ax.scatter(1, 1, 0) # z
ax.scatter(0, 1, 1) # w

plt.show()
```

### Задание 2.3.1

```python
iris = sns.load_dataset('iris')

x_train, x_test, y_train, y_test = train_test_split(
    iris.iloc[:, :-1],
    iris.iloc[:, -1],
    test_size=0.15
)

plt.figure(figsize=(16, 7))

# k = 1
model = KNeighborsClassifier(n_neighbors=1)
model.fit(x_train, y_train)

sns.scatterplot(x='petal_width', y='petal_length', data=iris, hue='species', s=70)
plt.xlabel('Длина лепестнка, см')
plt.ylabel('Ширина лепестка, см')
plt.legend()
plt.grid()

y_pred = model.predict(x_test)

for i in range(len(y_test)):
    if np.array(y_test)[i] != y_pred[i]:
        plt.scatter(x_test.iloc[i, 3], x_test.iloc[i, 2], color = 'red', s = 150)

print(f'accuracy (k=1): {accuracy_score(y_test, y_pred) : .3}')
```

```python
# k = 5
model = KNeighborsClassifier(n_neighbors=5)
model.fit(x_train, y_train)

plt.figure(figsize=(16, 7))

sns.scatterplot(x='petal_width', y='petal_length', data=iris, hue='species', s=70)
plt.xlabel('Длина лепестнка, см')
plt.ylabel('Ширина лепестка, см')
plt.legend()
plt.grid()

y_pred = model.predict(x_test)

for i in range(len(y_test)):
    if np.array(y_test)[i] != y_pred[i]:
        plt.scatter(x_test.iloc[i, 3], x_test.iloc[i, 2], color = 'red', s = 150)

print(f'accuracy (k=5): {accuracy_score(y_test, y_pred) : .3}')
```

```python
# k = 10
model = KNeighborsClassifier(n_neighbors=10)
model.fit(x_train, y_train)

plt.figure(figsize=(16, 7))

sns.scatterplot(x='petal_width', y='petal_length', data=iris, hue='species', s=70)
plt.xlabel('Длина лепестнка, см')
plt.ylabel('Ширина лепестка, см')
plt.legend()
plt.grid()

y_pred = model.predict(x_test)

for i in range(len(y_test)):
    if np.array(y_test)[i] != y_pred[i]:
        plt.scatter(x_test.iloc[i, 3], x_test.iloc[i, 2], color = 'red', s = 150)

print(f'accuracy (k=10): {accuracy_score(y_test, y_pred) : .3}')
```
