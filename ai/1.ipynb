### Задание 4.3.5

```python
def plus(a, b):
    return a + b 

def minus(a, b):
    return a - b 

def multiply(a, b):
    return a * b 

def divide(a, b):
    return a / b 

def fe(x, y):
    return e**(x + y)

def fsin(x, y):
    return sin(x + y)

def fcos(x, y):
    return cos(x + y)

def pow(x, y):
    return x ** y

while True:
    s = input("Enter (+,-,*,/,e^(x+y),sin(x+y),cos(x+y),x^y) or stop: ")
    if s == 'stop':
        break
    if s in ('+', '-', '*', '/', 'e^(x+y)', 'sin(x+y)', 'cos(x+y)', 'x^y'):
        x = float(input("x="))
        y = float(input("y="))
        if s == '+':
            print("%.2f" % plus(x, y))
        elif s == '-':
            print("%.2f" % minus(x, y))
        elif s == '*':
            print("%.2f" % multiply(x, y))
        elif s == '/':
            if y != 0:
                print("%.2f" % divide(x, y))
            else:
                print("dividing by zero!")
        elif s == 'e^(x+y)':
            print("%.2f" % fe(x, y))
        elif s == 'sin(x+y)':
            print("%.2f" % fsin(x, y))
        elif s == 'cos(x+y)':
            print("%.2f" % fcos(x, y))
        elif s == 'x^y':
            print("%.2f" % pow(x, y))
```
